```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fat_df = pd.read_csv('C:/Users/82103/Downloads/archive/Fat_Supply_Quantity_Data.csv')
protein_df = pd.read_csv('C:/Users/82103/Downloads/archive/Protein_Supply_Quantity_Data.csv')
```

```python
fat_df.isna().sum()
```

```python
fat_df.dropna(axis = 0, how = 'any', inplace = True)
protein_df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
def change(x):
    if x == '<2.5':
        x = 0
        return float(x)
    else:
        return float(x)
```

```python
fat_df['Undernourished'] = fat_df.Undernourished.apply(lambda x: change(x))
protein_df['Undernourished'] = protein_df.Undernourished.apply(lambda x: change(x))
```

```python
def correlation_heatmap(df,size=(30,30)):
    _ , ax = plt.subplots(figsize = size)
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':9 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)
```

```python
correlation_heatmap(fat_df,size=(30,30))
```
